```python
import numpy as np 
import pandas as pd 
```

```python
# merge two dataframes 

employees =pd.DataFrame({
    "employee_id":[1,2,3,4,5],
    "NAME":["CHETAN ", "HARI","RAM","ROHAN","RICHA"],
    "DEPARTMENT":["HR","IT","HR","IT","FINANCE"]
    
})
SALARIES = pd.DataFrame({
    "employee_id":[1,2,3,6,7],
    "salary":[6000,5000,460000,30000,20000],
    "bonus":[2000,3000,2000,1000,6000]
})
```

```python
employees
```

```python
SALARIES
```

```python
pd.merge(employees,SALARIES,on="employee_id",how="inner")
```

```python
pd.merge(employees,SALARIES,on="employee_id",how="left")
```

```python
pd.merge(employees,SALARIES,on="employee_id",how="right")
```

```python
pd.merge(employees,SALARIES,on="employee_id",how="outer")
```

```python
df1 =pd.DataFrame({
    "A":["A0","A1","A2"],
    "B":["B0","B1","B2"],
    "C":["C0","C1","C2"]
    
})
df2 = pd.DataFrame({
    "A":["A3","A4","A5"],
    "B":["B3","B4","B5"],
    "C":["C3","C4","C5"]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1 =pd.DataFrame({
    "name":["heri","linda","charlie"]
},index=[1,2,3])

df2 =pd.DataFrame({
    "score":[85,78,98]
},index=[2,3,4])
```
