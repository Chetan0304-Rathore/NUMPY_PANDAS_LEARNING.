```python
import numpy as np
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [50000, 70000, 55000, 80000, 60000]
}

df = pd.DataFrame(data)
df
```

```python
# group by department and calculate the sum of salary
df.groupby("Department")["Salary"].sum()
```

```python
# group by emlpoyee and calculate the sum of salary 
df.groupby("Employee")["Salary"].sum()
```

```python
df2 = df.groupby(["Department","Employee"])["Salary"].sum()
df2
```
